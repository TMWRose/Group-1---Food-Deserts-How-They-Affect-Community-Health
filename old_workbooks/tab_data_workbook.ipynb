```python
import pandas as pd
```

```python
### Read in the data from CSV as dataframe

df=pd.read_csv('FoodEnvironmentAtlas_proposed.csv')
df=df.iloc[:, 1:]
```

```python
### Drop the rows with missing values from the end of the file

df=df.drop(index=[3142, 3143])
```

```python
## Database Dependencies and configuration path

import sqlalchemy as sql
import config
```

```python
# Create engine to connect to the database

endpoint=config.aws_endpoint
username='postgres'
password=config.aws_password
engine=sql.create_engine(f'postgresql://{username}:{password}@{endpoint}:5432/postgres')
```

```python
# Write the datarame to the database with the connection we created above

df.to_sql('postgres', con=engine, index=False, if_exists='replace')
```

```python
# Check the database to see if the data was written
df=pd.read_sql_table('postgres', con=engine)
```
